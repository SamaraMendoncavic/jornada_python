```python
# Importando bibliotecas necessárias
import pandas as pd

# Para converter texto em número
from sklearn.preprocessing import LabelEncoder

# Para pegar o X e Y, separa entre treino e teste
from sklearn.model_selection import train_test_split

# Importação de modelos de IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Impostação do cálculo da acuracia
from sklearn.metrics import accuracy_score
```

```python
# Iniciando o código
tabela = pd.read_csv("clientes.csv")
novos_clientes = pd.read_csv("novos_clientes.csv")

# Exibir informações da tabela
display(tabela)
display(tabela.info())
```

**Passo 2:** Tratar e Preparar a base de dados para a inteligência artificial

```python
# Desafio: Prever a coluna "score_credito"

# Transformar todas as colunas que são textos em números
# Profissao
codificador_prof = LabelEncoder()
tabela['profissao'] = codificador_prof.fit_transform(tabela["profissao"])

# mix_credito
codificador_mix = LabelEncoder()
tabela['mix_credito'] = codificador_mix.fit_transform(tabela['mix_credito'])

# comportamento_pagamento
codificador_pag = LabelEncoder()
tabela['comportamento_pagamento'] = codificador_pag.fit_transform(tabela['comportamento_pagamento'])

display(tabela.info())
```

**Como funciona o processo de criação da IA?**





```python
# Y = Quem eu quero prever
y = tabela['score_credito']

# X = Quem eu vou usar para fazer a previsão
x = tabela.drop(columns=["score_credito", "id_cliente"])

# Vai criar separação para realizar o teste e o treino para a IA
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

**Passo 3:** Modelo de previsão ou modelo de IA -> Definir nota do cliente

**Modelos:**
1. Árvore de Decisão -> RandomForest
2. Vizinhos Próximos -> KNN -> Nearest Neighbors

```python
# Importar a IA

# Criar a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinar a IA
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

**Passo 4:** Avaliar e Escolher o melhor modelo

```python
# Previsão
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Comparar a previsão com o y_teste
print("Previsão Árvore: ", accuracy_score(y_teste, previsao_arvoredecisao))
print("Previsão KNN: ", accuracy_score(y_teste, previsao_knn))
```

Passo que podem ajudar a melhor a precisão do modelo:

**Passo 5:** Fazer novas previsões

Melhor modelo (Neste caso) = Árvore de Decisão

```python
# Tabela de nosvos clientes
display(novos_clientes.info())

# Transformar todas as colunas que são textos em números
# Profissao
codificador_prof = LabelEncoder()
novos_clientes['profissao'] = codificador_prof.fit_transform(novos_clientes["profissao"])

# mix_credito
codificador_mix = LabelEncoder()
novos_clientes['mix_credito'] = codificador_mix.fit_transform(novos_clientes['mix_credito'])

# comportamento_pagamento
codificador_pag = LabelEncoder()
novos_clientes['comportamento_pagamento'] = codificador_pag.fit_transform(novos_clientes['comportamento_pagamento'])

print("Transformação dos dados feita com sucesso!")

previsao = modelo_arvoredecisao.predict(novos_clientes)

display(novos_clientes.info())
display(previsao)
```
